```python
import cv2

def get_qr_code_image(qr_code):
    img = cv2.imread(qr_code, 0)
    cv2.imshow('a', img)

    return cv2.QRCodeDetector().detectAndDecode(img)

val, pts, st_code = get_qr_code_image(r"qr_code/chunk_3.png")
print('#'*20, type(val), len(val))
print(val)
```
